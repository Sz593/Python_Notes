# Basic Graphing with Matplotlib

```python
%matplotlib inline
```

The above line is an IPython notebook magic command. This allows us to see any plots and graphs we create inline, instead of creating them in separate pop-up windows. 

We start by importing pylab and using the basic plot function.

```python
# import pylab
import pylab

# Create a list of values to plot
x_range = [0,1,2,3,4]
y_range = [4,5,6,7,8]
```

To create a graph, we call the plot function:

```python
pylab.plot(x_range, y_range)
```

Normally, the plot function itself doesn't bring up the graph, it just creates it. To see the graph, we need to call the show() function. However, since we used the %matplotlib inline command, the plots show up in the IPython notebook.

We can add a marker to this graph as well:

```python
pylab.plot(x_range, y_range, marker='o')
```

There are several options for markers, including: 'o', '*', 'x', and '+'. The "marker=" section indicates that the markers should be connected by a line. Leaving this off removes the line. 

```python
pylab.plot(x_range, y_range, 'o')
```

###Graphing the Average Annual Temperature in NYC

```python
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
pylab.plot(nyc_temp, marker='o')
```

When you call plot() on a single list, the values are plotted on the y-axis against the list indices on the x-axis. We can label the x-axis by creating another list called years and plotting this against the temperatures. 

```python
years = range(2000,2013)
pylab.plot(years, nyc_temp, marker='o')
```

###Comparing the Monthly Temperature Trends of NYC

If we want to compare the average monthly temperatures of three different years on the same graph, we can do that. 

```python
# Create lists of temperatures
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]

# create list of months
months = range(1,13)
```

The clearest way to compare these will be to plot them on the same graph. Each plot needs to be a pair of lists. Matplotlib will create three pair of plots, but will display them on top of each other. 

```python
pylab.plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
```

This is a great start, but we don't know which color corresponds to which year. Ergo, we have to add a legend.

```python
pylab.plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
pylab.legend([2000,2006,2012])
```

We can also customize the axes:

```python
pylab.plot(years, nyc_temp, marker='o')
pylab.axis()
```

The pylab.axis() function returns two tuples: the x-axis range and the y-axis range in the form (xmin, xmax, ymin, ymax). You can manually set some of these values here. 

```python
pylab.plot(years, nyc_temp, marker='o')
pylab.axis(ymin=0)
pylab.title('Average Temperature in NYC')
pylab.xlabel('Year')
pylab.ylabel('Temperature in Fahrenheit')
```

Pylab was created for creating plots in the interactive shell, but when plotting in larger programs, you should use pyplot.

###Plotting Newton's Law of Universal Gravitation

According to Newton's law of gravitation, a body of mass $m_1$ is attracted to another body of mass $m_2$ with a force $F$ according to
$$ F = \frac{Gm_1m_2}{r^2} $$
where $r$ is the distance between the two bodies, and $G$ is the universal gravitational constant, approximately $6.674 \times 10^{-11}$.

We can use this formula to calculate $F$ at several different distances. 

```python
# Define distances, F, and constants.
r = range(100,1001,50)
G = 6.674*(10**-11)
```

```python
# Prompt user for masses

while True:
    try:
        m1 = int(input('Please enter the first mass.'))
        m2 = int(input('Please enter the second mass.'))
        break
    except ValueError:
        print('Please enter an integer value for each mass.')
```

```python
F = map(lambda x: (G * m1 * m2) / x**2, r)
list_F = list(F)
```

```python
plt.plot(r, list_F, marker='o')
```

###Bar Charts

This is some sample code for creating a bar horizontal bar chart, based on the number of steps I walked in the past week.

```python
# Number of steps I walked during the past week
steps = [6534, 7000, 8900, 10786, 3467, 11045, 5095]
# Corresponding days
labels = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']

# Number of bars
num_bars = len(steps)

# This list is the point on the y-axis where each
# Bar is centered. Here it will be [1, 2, 3...]
positions = range(1, num_bars+1)
plt.barh(positions, steps, align='center')

# Set the label of each bar
plt.yticks(positions, labels)
plt.xlabel('Steps')
plt.ylabel('Day')
plt.title('Number of steps walked')
# Turns on the grid which may assist in visual estimation
plt.grid()
```

###Pyplot vs. Pylab

Matplotlib is the whole package; pylab is a module in matplotlib that gets installed alongside matplotlib; and matplotlib.pyplot is a module in matplotlib.

Pyplot provides the state-machine interface to the underlying plotting library in matplotlib. This means that figures and axes are implicitly and automatically created to achieve the desired plot. For example, calling plot from pyplot will automatically create the necessary figure and axes to achieve the desired plot. Setting a title will then automatically set that title to the current axes object:

Pylab combines the pyplot functionality (for plotting) with the numpy functionality (for mathematics and for working with arrays) in a single namespace, making that namespace (or environment) even more MATLAB-like. For example, one can call the sin and cos functions just like you could in MATLAB, as well as having all the features of pyplot.

The pyplot interface is generally preferred for non-interactive plotting (i.e., scripting). The pylab interface is convenient for interactive calculations and plotting, as it minimizes typing. Note that this is what you get if you use the ipython shell with the -pylab option, which imports everything from pylab and makes plotting fully interactive.
