```python
# standard imports
import numpy as np
import pandas as pd

# stats import
from scipy import stats

# plotting imports
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# inline graphing
%matplotlib inline

# suppress jupyter notebook warnings
import warnings
warnings.filterwarnings('ignore')
```

Seaborn comes with some example datasets. Let's use the tips sample data set for these exercises. 

```python
# view the list of sample pandas datasets from seaborn
sns.get_dataset_names()
```

```python
# save the tips dataset to a variable
tips = sns.load_dataset('tips')
```

```python
# see the first 5 rows
tips.head()
```

```python
# let's start by plotting the tips vs. the total bill
sns.lmplot('total_bill', 'tip', tips)
```

We can see a scatter plots for all the data, overlayed with a linear regression. 

```python
# create the regression plot with some options specified
sns.lmplot('total_bill', 'tip', tips, 
           scatter_kws = {'marker':'o', 
                          'color':'indianred'},
           line_kws = {'linewidth':1,
                       'color':'blue'})
```

```python
# change the order of the polynomial fit (linear = 1 to polynomial)
sns.lmplot('total_bill', 'tip', tips, order = 4,
           scatter_kws = {'marker':'o', 
                          'color':'indianred'},
           line_kws = {'linewidth':1,
                       'color':'blue'})
```

```python
# we can also just make a pure scatter plot
sns.lmplot('total_bill', 'tip', tips, fit_reg = False)
```

We can also plot discrete varibles, like the % of the tip.

```python
# create a new column in the dataframe called tip_percent
tips['tip_percent'] = 100 * tips['tip'] / tips['total_bill']
tips.head()
```

```python
# plot the new variable
sns.lmplot('size', 'tip_percent', tips)
```

To make this easier to read, we can add a [jitter](https://en.wikipedia.org/wiki/Jitter) to it. This spreads out our data a little bit and makes it easier to read. 

```python
# plot this again with a slight jitter on the x-axis
sns.lmplot('size', 'tip_percent', tips, x_jitter = 0.1)
```

```python
# measure the tendency of each bin
sns.lmplot('size', 'tip_percent', tips, x_estimator = np.mean)
```

```python
# use the hue to define substratify the data
sns.lmplot('total_bill', 'tip_percent', tips,
           hue = 'sex', markers = ['x', 'o'])
```

```python
# plot this again the the hue by day
sns.lmplot('total_bill', 'tip_percent', tips, hue = 'day')
```

Seaborn can also do [local regression](https://en.wikipedia.org/wiki/Local_regression) (LOWESS) model fitting. Local regression is a non-parametric regression method that combine multiple regression models in a k-nearest-neighbor-based meta-model.

It combines much of the simplicity of linear least squares regression with the flexibility of nonlinear regression. It does this by fitting simple models to localized subsets of the data to build up a function that describes the deterministic part of the variation in the data, point by point. In fact, one of the chief attractions of this method is that the data analyst is not required to specify a global function of any form to fit a model to the data, only to fit segments of the data.

```python
# create a local regrssion model
sns.lmplot('total_bill', 'tip_percent', tips,
           lowess = True,
           line_kws = {'color':'blue'},
           scatter_kws = {'color':'indianred'})
```

```python
# seaborn lmplot was built on regplot
# let's take a look at the lower-level plot
sns.regplot('total_bill', 'tip_percent', tips)
```

```python
# plot two figures next to each other
# sharey = True means they share a single y-axis
fig, (axis1, axis2) = plt.subplots(1, 2, sharey = True)

sns.regplot('total_bill', 'tip_percent', tips, ax = axis1)
sns.violinplot(x = 'size', y = 'tip_percent',
               data = tips,
               palette = 'Set3',
               ax = axis2)
```
