#Array Processing

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We're going to make a grid. We'll start be defining all point values, and then translating that into the grid.

```python
# Create array of possible point values
points = np.arange(-5,5,0.1)
```

```python
# Use meshgrid to return coordinate matrices from coordinate vectors
dx, dy = np.meshgrid(points, points)
```

```python
# Notice how each row goes from -5 to 5
dx
```

```python
# For dy, each column goes from -5 to 5
dy
```

```python
# Take the sine of both dx and dy
z = (np.sin(dx) + np.sin(dy))
```

```python
# Plot the sine values stored in z using plt.imshow(). 
# Image show (imshow()) tends to be a good way to look at matrix data
plt.imshow(z)
```

```python
# To see more detail about this image, we'll add a colorbar and title
plt.imshow(z)
plt.colorbar()
plt.title('Plot of Sin(x) + Sin(y)')
```

###Numpy Where

```python
# Create two arrays
a = np.array([1,2,3,4])
b = np.array([100,200,300,400])
```

```python
# Create a condition array
condition = np.array([True, True, False, False])
```

For the next bit of code, we'll be using the built-in zip function. Here's a quick refresher on the zip function.

```python
# Zip function refresher: start by creating two lists
testlist1 = [1,2,3,4]
testlist2 = [5,6,7,8]

# zip the lists together
testzip = zip(testlist1, testlist2)

# print the zip values
for ziptuple in testzip:
    print(ziptuple)
```

Now we can return to numpy.

```python
# List comprehension
answer = [(A if cond else B) for A, B, cond in zip(a, b, condition)]
print(answer)
```

```python
# Using Numpy where
answer2 = np.where(condition, a, b)
print(answer2)
```

Using Numpy where, we can create arrays in a similar fashion to conditional list comprehensions. In Numpy where, the condition comes first, and then the two input arguments. 

```python
# Create random normal array c
c = np.random.randn(5,5)
print(c)
```

```python
# Numpy where, change all values below zero to zero
np.where(c < 0, 0, c)
```

###Statistical Functions on Matrices

```python
d = np.arange(1,10).reshape(3,3)
print(d)
```

```python
# Sum all values in the matrix
d.sum()
```

```python
# Sum up columns (axis = 0 for columns, axis = 1 for rows)
d.sum(0)
```

```python
# Mean value in matrix
d.mean()
```

```python
# Standard deviation in matrix
d.std()
```

```python
# Variance in matrix
d.var()
```

#Array I/O

```python
# Create an array
array1 = np.arange(5)
print(array1)
```

We can save arrays using np.save(), and then load them back using np.load(). The arrays are saved using a .npy file extension, and you can choose the name to save them as. 

```python
# Save array1
np.save('test_save_array', array1)
```

To show how save works, we're going reassign the variable array1 to a new array.

```python
# Create a new array1
array1 = np.random.randint(0,10,5)
print(array1)
```

So now that array1 has been asigned a new value, let's see if we can load the saved values into a new variable. Note that to load a saved array, we need the .npy extension.

```python
# Load the saved array into array2
array2 = np.load('test_save_array.npy')
print(array2)
```

Thus, we can see the original saved values were loaded into array2.

We can also save multiple arrays into a numpy zip file, using the savez() method. We currently have two different arrays, array1 and array2. To save these together, we can use a zip file. Note that a numpy zip file has a .npz extension, instead of .npy.

```python
# Save array1 and array2 to a zip file
np.savez('test_save_zip', saved_array1 = array1, saved_array2 = array2)
```

Once we have the arrays saved, we can load them back. 

```python
# Load the saved zip file into the archived_array variable
archived_array = np.load('test_save_zip.npz')
```

To access these loaded arrays, we call the variable archived_array with the name of the array we want to load as an argument. 

```python
# Call the first archived array
archived_array['saved_array1']
```

```python
# Call the second archived array
archived_array['saved_array2']
```

We can also save arrays as text files instead.

```python
# Create a matrix
array3 = np.arange(1,7).reshape(2,3)
print(array3)
```

We can save this matrix as a .txt file and specify the delimiter.

```python
# Save array3 as a text file
np.savetxt('text_save_text.txt', array3, delimiter=',')
```

```python
# Load the saved .txt file into array4
array4 = np.loadtxt('text_save_text.txt', delimiter=',')
print(array4)
```

Just as a note, the saved .npy, .npz, and .txt files are actually saved files on your disk. That means if you end your python session, you can load them the next time you open python.
