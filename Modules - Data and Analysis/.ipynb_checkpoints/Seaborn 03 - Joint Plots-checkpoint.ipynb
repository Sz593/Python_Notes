```python
# standard imports
import numpy as np
import pandas as pd

# statistics import
from scipy import stats

# import plotting packages/modules
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# display graphs inline
%matplotlib inline
```

```python
# make a dataset
dataset = np.random.randn(100)
```

```python
# we can combine plots using seaborn's distplot
sns.distplot(dataset, bins = 25)
```

```python
# we can turn different plots on and off with .distplot() options
sns.distplot(dataset, bins = 25, rug = True, hist = False)
```

```python
# we can use dictionaries to specify options for the underlying plots
sns.distplot(dataset, bins = 25,
             kde_kws = {'color':'indianred', 'label':'KDE'},
             hist_kws = {'color':'blue', 'label':'Histogram'})
```

```python
ser1 = pd.Series(dataset, name = 'Data Set 1')
ser1.head()
```

```python
# plot a pandas series
sns.distplot(ser1, bins = 25)
```
