```python
import numpy as np
import pandas as pd
```

```python
# go to a website to download data
import webbrowser
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
webbrowser.open(url)
#download winequality-red.csv
```

```python
# create dataframe using ; as the delimiter
df_wine = pd.read_csv('winequality-red.csv', sep = ';')
df_wine.head()
```

```python
# find average alcohol content for the wines
df_wine['alcohol'].mean()
```

```python
# create our own aggregate function
def var_range(df):
    return df.max() - df.min()
```

```python
# quality rating
wine_quality_group = df_wine.groupby('quality')
wine_quality_group.describe()
```

```python
# use the .agg() method to pass user-created aggregate functions
wine_quality_group.agg(var_range)
```

```python
# you can also use the .agg() method to calcualte standard aggregations
wine_quality_group.agg('mean')
```

Let's determine if the alcohol quality has a link to the percent alcohol of the wine. 

```python
# create a new column for quality to alcohol ratio
df_wine['qual/alc ratio'] = df_wine['quality'] / df_wine['alcohol']
df_wine.head()
```

```python
# alternatively, we could use a pivot table pivoted on the quality column
df_wine_pivot = df_wine.pivot_table(index = 'quality')
df_wine_pivot
```

```python
# prepare to visualize the data
%matplotlib inline
```

```python
# create a scatter plot
df_wine.plot(kind = 'scatter', x = 'quality', y = 'alcohol')
```

```python
df_wine_pivot.plot(kind = 'barh', y = 'alcohol', 
                   title = 'Quality by Alcohol Content',
                   legend = False)
```
