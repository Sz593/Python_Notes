```python
# use the red wine data we used in the previous Pandas Data 8 notebook
df_wine = pd.read_csv('winequality-red.csv', sep = ';')
df_wine.head()
```

```python
# create a function to rank each wine by alcohol content
def rank_wine(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
# sort data frame by alcohol content in descending order
df_wine.sort_values(by = 'alcohol', ascending = False, inplace = True)
df_wine.head(10)
```

```python
# apply rank_wine function in for each quality group
df_wine = df_wine.groupby('quality').apply(rank_wine)
df_wine.head(5)
```

```python
# find the number of wines in each quality
num_of_quality = df_wine['quality'].value_counts()
num_of_quality
```
