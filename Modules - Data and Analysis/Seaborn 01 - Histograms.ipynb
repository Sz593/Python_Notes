```python
# standard imports
import numpy as np
import pandas as pd

# statistics imports
from scipy import stats

# plotting imports 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# create a dataset of 100 normally distributed values
dataset1 = np.random.randn(100)
```

```python
# make a histogram of this data
plt.hist(dataset1)
```

```python
# make another data set
dataset2 = np.random.randn(80)
# plot a histogram of the dataset
plt.hist(dataset2, color = 'indianred')
```

```python
# show both histograms on the same plot
# alpha is the transparency, normed normalizes the data sets
plt.hist(dataset1, normed = True, alpha = 0.5, bins = 20)
plt.hist(dataset2, normed = True, color = 'indianred', alpha = 0.5, bins = 20)
```

```python
# create two more data sets
dataset3 = np.random.randn(1000)
dataset4 = np.random.randn(1000)
```

```python
# seaborn joint plot
sns.jointplot(dataset3, dataset4)
```

```python
sns.jointplot(dataset3, dataset4, kind = 'hex')
```
